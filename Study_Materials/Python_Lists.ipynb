### What are Lists

A list is a data structure in Python that is a mutable, or changeable, ordered sequence of elements. Each element or value that is inside of a list is called an item. Just as strings are defined as characters between quotes, lists are defined by having values between square brackets [ ]. Lists act like dynamic arrays which means that can add more items on the fly.

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(5)
```

```python
# index
L = [1,2,1,3,4,1,5]
L.index(1)
```

```python
# reverse
L = [2,1,5,7,0]
# permanently reverses the list
L.reverse()
print(L)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# copy -> shallow
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
L = [5*i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]
#x=[]
#for i in v:
#  x.append(i*s)
#print(x)
[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
#[i for i in range(0,51)]
[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('j')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,5)]
```

```python
# A products like multiplication-> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```
